```python
import requests

url = "https://strapi.tungshin.com.my/graphql"

payload = {
    "operationName": "DoctorsBySpecialtiesByName",
    "variables": {
        "limit": 200,
        "start": 0,
        "sort": "actualname:asc",
        "specialtiesId": [1, 7, 5, 9, 10, 13, 8, 11, 16, 18, 19, 12, 17, 14, 15, 20, 22, 23, 24, 25, 26, 35, 43, 44],
        "name": None,
        "namecn": None
    },
    "query": "query DoctorsBySpecialtiesByName($limit: Int, $start: Int, $sort: String, $name: String, $namecn: String, $specialtiesId: [ID]) {\n  doctors(\n    limit: $limit\n    start: $start\n    sort: $sort\n    where: {_or: [{name_contains: $name}, {name_cn_contains: $namecn}], specialties: {id_in: $specialtiesId}}\n  ) {\n    id\n    name\n    name_cn\n    title\n    actualname\n    slug\n    flip\n    placement\n    location\n    location_zh\n    location_my\n    contact\n    portrait {\n      url\n      formats\n      caption\n      __typename\n    }\n    specialties {\n      id\n      name\n      name_cn\n      name_bm\n      __typename\n    }\n    doctor_hours {\n      opening_hours\n      day_interval\n      __typename\n    }\n    doctor_qualifications {\n      year\n      qualification\n      __typename\n    }\n    doctor_languages {\n      language\n      __typename\n    }\n    __typename\n  }\n  totalCount: doctorsConnection(\n    where: {name_contains: $name, specialties: {id_in: $specialtiesId}}\n  ) {\n    aggregate {\n      count\n      __typename\n    }\n    values {\n      id\n      __typename\n    }\n    __typename\n  }\n}\n"
}

response = requests.post(url, json=payload)

if response.status_code == 200:
    data = response.json()
    # Process the data as needed
else:
    print(f"Request failed with status code {response.status_code}")
```
